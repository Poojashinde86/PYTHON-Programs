```python
import nltk
from nltk.chat.util import Chat, reflections
```

```python
#reflections which is a dictionary containing basic input and corresponding outputs. 

reflections = {
  "i am"       : "you are",
  "i was"      : "you were",
  "i"          : "you",
  "i'm"        : "you are",
  "i'd"        : "you would",
  "i've"       : "you have",
  "i'll"       : "you will",
  "my"         : "your",
  "you are"    : "I am",
  "you were"   : "I was",
  "you've"     : "I have",
  "you'll"     : "I will",
  "your"       : "my",
  "yours"      : "mine",
  "you"        : "me",
  "me"         : "you"
}
```

```python
#create a simple set of rules. 
#The first line describes the user input which we have taken as raw string input and the next line is our chatbot response.

pairs = [
    [
        r"my name is (.*)",
        ["Hello %1, How are you today ?",]
    ],
    [
        r"hi|hey|hello",
        ["Hello", "Hey there",]
    ], 
    [
        r"what is your name ?",
        ["I am a bot created using python. You can call me anything!",]
    ],
    [
        r"how are you ?",
        ["I'm doing good and How about You?",]
    ],
    [
        r"sorry (.*)",
        ["Its alright","Its OK, never mind",]
    ],
    [
        r"I am fine",
        ["Great to hear that, How can I help you?",]
    ],
    [
        r"i'm (.*) doing good",
        ["Nice to hear that","How can I help you?:)",]
    ],
    [
        r"(.*) age?",
        ["I'm a computer program .Seriously you are asking me this?",]
    ],
    [
        r"what (.*) want?",
        ["Make me an offer I can't refuse",]
    ],
    [
        r"(.*) created ?",
        ["Team created me using Python's NLTK library ","top secret ;)",]
    ],
    [
        r"(.*) (location|city) ?",
        ['Pune, Maharashtra',]
    ],
    [
        r"how is weather in (.*)?",
        ["Too hot man here in %1","Too hot outside!!! ,Temp is around 45 degrees",]
    ],
    [
        r"i work in (.*) ?",
        ["%1 is an Amazing company, I have heard about it. But they are in huge loss these days.",]
    ],
    
    [
        r"how (.*) health(.*)",
        ["I'm a computer program, so I'm always healthy ",]
    ],
    [
        r"(.*) (sports|game) ?",
        ["I'm a very big fan of Football",]
    ],
    [
        r"who (.*) sportsperson ?",
        ["Messy","Ronaldo","Roony"]
    ],
    [
        r"who (.*) (moviestar|actor)?",
        ["Brad Pitt"]
    ],
    [
        r"i am looking for online guides and courses to learn data science, can you suggest?",
        ["Statquest has many great articles with each step explanation along with code, you can explore"]
    ],
     [
        r"i have query?",
        ["How can I help you","Always there to help, please ask"]
    ],
     [
        r"(.*)getting bored?",
        ["you can call your friends and go outside","Hang out with friends","do whatever is your hobby, you will feel better"]
    ],
     [
        r"how can i learn java?",
        ["W3Shool is great option, Javatpoint is good for theory"]
    ],
    [
        r"bye",
        ["BBye take care. See you soon :) ","It was nice talking to you. See you soon :)"]
    ],
]
```

```python
#define a function to initiate the chat process.  
#And the conversation starts from here by calling a Chat class and passing pairs and reflections to it.

def chat():
    print("Hi! I am a chatbot created using python for your service")
    chat = Chat(pairs, reflections)
    chat.converse()
#initiate the conversation
if __name__ == "__main__":
    chat()
```
